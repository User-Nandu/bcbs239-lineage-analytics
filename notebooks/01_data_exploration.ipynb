```python
from neo4j import GraphDatabase

uri = "bolt://localhost:7687"
username = "neo4j"
password = "cn4125555"

driver = GraphDatabase.driver(uri, auth=(username, password))

def create_source(tx):
    tx.run("CREATE (:SourceSystem {name: 'TradingSystem'})")

with driver.session() as session:
    session.execute_write(create_source)

print("Source system node created")
```

```python
def create_etl_job(tx):
    tx.run("""
    MATCH (s:SourceSystem {name:'TradingSystem'})
    CREATE (e:ETLJob {name:'DailyRiskETL'})
    CREATE (s)-[:FEEDS]->(e)
    """)

with driver.session() as session:
    session.execute_write(create_etl_job)

print("ETL job node created")
```

```python
def create_transformation(tx):
    tx.run("""
    MATCH (e:ETLJob {name:'DailyRiskETL'})
    CREATE (t:Transformation {name:'AggregateRiskMetrics'})
    CREATE (e)-[:TRANSFORMS]->(t)
    """)

with driver.session() as session:
    session.execute_write(create_transformation)

print("Transformation node created")
```

```python
def create_target_table(tx):
    tx.run("""
    MATCH (t:Transformation {name:'AggregateRiskMetrics'})
    CREATE (target:TargetTable {name:'RiskAggregationTable'})
    CREATE (t)-[:LOADS_INTO]->(target)
    """)

with driver.session() as session:
    session.execute_write(create_target_table)

print("Target table node created")
```

```python
def create_risk_report(tx):
    tx.run("""
    MATCH (target:TargetTable {name:'RiskAggregationTable'})
    CREATE (report:RiskReport {name:'BaselRiskReport'})
    CREATE (target)-[:USED_IN]->(report)
    """)

with driver.session() as session:
    session.execute_write(create_risk_report)

print("Risk report node created")
```
